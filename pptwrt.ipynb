```
from openpyxl import load_workbook
from pptx import Presentation

# Load the workbook and get the active sheet
wb = load_workbook('presentation_data.xlsx')
ws = wb.active

# Load the presentation
presentation = Presentation('CONTOSO.pptx')

# Iterate over the rows in the Excel file
for row in ws.iter_rows(min_row=2, values_only=True):  # start from 2 to skip header
    slide_num, object_name, object_type, text = row

    # Get the slide and the shape
    slide = presentation.slides[slide_num - 1]
    shape = next((s for s in slide.shapes if s.name == object_name), None)

    # Check if the shape was found and it has text
    if shape is not None and shape.has_text_frame:
        # Check if text is not None
        if text is not None:
            # Set the text of the shape
            shape.text = text
        else:
            # Clear the text of the shape
            shape.text = ''

# Save the presentation
presentation.save('CONTOSO2.pptx')
```
