```
from pptx import Presentation
from openpyxl import Workbook

# Load the presentation
presentation = Presentation('CONTOSO.pptx')

# Create a new workbook and get the active sheet
wb = Workbook()
ws = wb.active

# Write the header row
ws.append(['Slide Number', 'Object Name', 'Object Type', 'First 3 Words'])

# Iterate over the slides
for slide_num, slide in enumerate(presentation.slides, start=1):
    # Iterate over the shapes in the slide
    for shape in slide.shapes:
        # Check if the shape has text
        if shape.has_text_frame:
            # Get the first 3 words
            text = ' '.join(shape.text.split()[:3])
            # Write the data to the Excel file
            ws.append([slide_num, shape.name, 'Text', text])
        else:
            # Write the data to the Excel file without text
            ws.append([slide_num, shape.name, 'Non-Text', ''])

# Save the workbook
wb.save('presentation_data.xlsx')
```
