```
import re
from pptx import Presentation
from openpyxl import Workbook

# Load the presentation
presentation = Presentation("CONTOSO.pptx")

# Create a new workbook and select the active sheet
wb = Workbook()
ws = wb.active

# Add headers to the sheet
ws.append(["Slide ID", "Text Box ID", "Text"])

# Initialize a counter for text box IDs
text_box_id = 1

# Define a function to remove illegal characters
def remove_illegal_chars(text):
    return re.sub(r'[\000-\010]|[\013-\014]|[\016-\037]', '', text)

# Iterate over the slides in the presentation
for slide_num, slide in enumerate(presentation.slides, start=1):
    # Iterate over the shapes in the slide
    for shape in slide.shapes:
        # Check if the shape is a text box
        if shape.has_text_frame:
            # Get the text in the text box
            text = "\n".join([p.text for p in shape.text_frame.paragraphs])
            # Remove illegal characters from the text
            text = remove_illegal_chars(text)
            # Add the slide number, text box id and text to the sheet
            ws.append([slide_num, text_box_id, text])
            # Increment the text box id
            text_box_id += 1

# Save the workbook
wb.save("text_boxes.xlsx")
```
